# TradeStation SDK - Authentication

This notebook demonstrates how to:
- Set up credentials
- Authenticate with TradeStation API
- Manage tokens
- Switch between PAPER and LIVE modes

## Prerequisites

1. TradeStation Developer account
2. Client ID and Client Secret from [TradeStation Developer Portal](https://developer.tradestation.com)
3. `.env` file with credentials (see below)

## Step 1: Install SDK

```bash
pip install tradestation-python-sdk
```

## Step 2: Set Up Environment Variables

Create a `.env` file in your project directory:

```env
TRADESTATION_CLIENT_ID=your_client_id_here
TRADESTATION_CLIENT_SECRET=your_client_secret_here
TRADESTATION_REDIRECT_URI=http://localhost:8888/callback
TRADING_MODE=PAPER
```

````python
# Import SDK
import os

from dotenv import load_dotenv
from tradestation_sdk import TradeStationSDK

# Load environment variables
load_dotenv()

# Verify credentials are loaded
print(f"Client ID: {os.getenv('TRADESTATION_CLIENT_ID')[:10]}...")
print(f"Trading Mode: {os.getenv('TRADING_MODE')}")
````

## Step 3: Initialize SDK

````python
# Initialize SDK
sdk = TradeStationSDK()

# Get SDK info
info = sdk.info()
print(f"SDK Version: {info['version']}")
print(f"API Version: {info['api_version']}")
print(f"Features: {list(info['features'].keys())}")
````

## Step 4: Authenticate (PAPER Mode)

This will open a browser window for TradeStation login (first time only).
After successful login, tokens are saved and authentication is automatic.

````python
# Authenticate with PAPER mode
try:
    sdk.authenticate(mode="PAPER")
    print("✅ Authentication successful!")

    # Check SDK info again
    info = sdk.info()
    print(f"Authenticated Modes: {info['authenticated_modes']}")
    print(f"Active Mode: {info['active_mode']}")
except Exception as e:
    print(f"❌ Authentication failed: {e}")
````

## Step 5: Test Authentication - Get Account Info

````python
# Get account information
account = sdk.get_account_info(mode="PAPER")

print("✅ Connected to account!")
print(f"Account ID: {account['account_id']}")
print(f"Account Name: {account['name']}")
print(f"Account Type: {account['type']}")
print(f"Status: {account['status']}")
print(f"Currency: {account['currency']}")
````

## Step 6: Get Account Balances

````python
# Get account balances
balances = sdk.get_account_balances(mode="PAPER")

print("💰 Account Balances:")
print(f"Equity: ${balances['equity']:,.2f}")
print(f"Cash Balance: ${balances['cash_balance']:,.2f}")
print(f"Buying Power: ${balances['buying_power']:,.2f}")
print(f"Margin Available: ${balances['margin_available']:,.2f}")
print(f"Margin Used: ${balances['margin_used']:,.2f}")
print(f"Open P&L: ${balances['open_pnl']:,.2f}")
````

## Step 7: Token Management

The SDK automatically refreshes tokens when they expire (20-minute lifespan).
You can also manually refresh or check token status.

````python
# Ensure authenticated (auto-refreshes if needed)
sdk.ensure_authenticated(mode="PAPER")
print("✅ Token is valid")

# Check authenticated modes
info = sdk.info()
print(f"Authenticated modes: {info['authenticated_modes']}")
````
